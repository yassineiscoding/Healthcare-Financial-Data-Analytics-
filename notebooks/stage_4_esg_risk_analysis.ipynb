## IV. ESG Risk Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import yesg
lly_esg = yesg.get_historic_esg('LLY')['2020-01':]
jnj_esg = yesg.get_historic_esg('JNJ')['2020-01':]
unh_esg = yesg.get_historic_esg('UNH')['2020-01':]
mrk_esg = yesg.get_historic_esg('MRK')['2020-01':]
```

This code will resample the original jnj_esg DataFrame to an annual frequency using the 'A' rule (for Annual), and it will calculate the mean value for each year.

```python
import pandas as pd
annual_esg = jnj_esg.resample('A').mean()
annual_esg.index = annual_esg.index.year
annual_esg
```

```python
# Process the LLY DataFrame
lly_esg.index = lly_esg.index.to_period('M')
lly_annual_esg = lly_esg.resample('A').mean()
lly_annual_esg.index = lly_annual_esg.index.year

# Process the JNJ DataFrame
jnj_esg.index = jnj_esg.index.to_period('M')
jnj_annual_esg = jnj_esg.resample('A').mean()
jnj_annual_esg.index = jnj_annual_esg.index.year

# Process the UNH DataFrame
unh_esg.index = unh_esg.index.to_period('M')
unh_annual_esg = unh_esg.resample('A').mean()
unh_annual_esg.index = unh_annual_esg.index.year

# Process the MRK DataFrame
mrk_esg.index = mrk_esg.index.to_period('M')
mrk_annual_esg = mrk_esg.resample('A').mean()
mrk_annual_esg.index = mrk_annual_esg.index.year

# Print the annual ESG DataFrames
print("Annual ESG Data for LLY:")
print(lly_annual_esg)
print("\n")

print("Annual ESG Data for JNJ:")
print(jnj_annual_esg)
print("\n")

print("Annual ESG Data for UNH:")
print(unh_annual_esg)
print("\n")

print("Annual ESG Data for MRK:")
print(mrk_annual_esg)
print("\n")
```

```python
# Create a list of years for the x-axis
years = list(lly_annual_esg.index)

# E-Score Comparison
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("E-Score Comparison")
ax.set_xlabel("Year")
ax.set_ylabel("Score")
sns.lineplot(x=years, y='E-Score', data=lly_annual_esg, marker='o', label='LLY', ax=ax)
sns.lineplot(x=years, y='E-Score', data=jnj_annual_esg, marker='o', label='JNJ', ax=ax)
sns.lineplot(x=years, y='E-Score', data=unh_annual_esg, marker='o', label='UNH', ax=ax)
sns.lineplot(x=years, y='E-Score', data=mrk_annual_esg, marker='o', label='MRK', ax=ax)
ax.legend()
plt.show()
```

**Environmental Performance (E-Score):**
   - LLY had the highest E-Score in 2022 (3.33) and the second-highest in 2020 (2.07), indicating relatively better environmental performance compared to the other companies.
   - JNJ and MRK had relatively low E-Scores throughout the period, with JNJ's E-Score dropping from 1.29 in 2020 to 0.92 in 2023, and MRK's E-Score fluctuating between 0.79 and 1.85.
   - UNH had a consistently low E-Score of 0.03 across all years, suggesting poor environmental performance.

```python
# S-Score Comparison
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("S-Score Comparison")
ax.set_xlabel("Year")
ax.set_ylabel("Score")
sns.lineplot(x=years, y='S-Score', data=lly_annual_esg, marker='s', label='LLY', ax=ax)
sns.lineplot(x=years, y='S-Score', data=jnj_annual_esg, marker='s', label='JNJ', ax=ax)
sns.lineplot(x=years, y='S-Score', data=unh_annual_esg, marker='s', label='UNH', ax=ax)
sns.lineplot(x=years, y='S-Score', data=mrk_annual_esg, marker='s', label='MRK', ax=ax)
ax.legend()
plt.show()
```

**Social Performance (S-Score):**
   - JNJ had the highest S-Score in 2020 (21.19), indicating better social performance compared to the other companies that year.
   - However, JNJ's S-Score declined from 21.19 in 2020 to 15.02 in 2023, suggesting a deterioration in social performance.
   - LLY and MRK had relatively stable S-Scores throughout the period, with LLY's S-Score ranging from 12.6 to 19.53, and MRK's S-Score ranging from 10.64 to 14.53.
   - UNH had the lowest S-Scores among the companies, ranging from 9.67 to 15.16, indicating weaker social performance.

```python
# G-Score Comparison
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("G-Score Comparison")
ax.set_xlabel("Year")
ax.set_ylabel("Score")
sns.lineplot(x=years, y='G-Score', data=lly_annual_esg, marker='^', label='LLY', ax=ax)
sns.lineplot(x=years, y='G-Score', data=jnj_annual_esg, marker='^', label='JNJ', ax=ax)
sns.lineplot(x=years, y='G-Score', data=unh_annual_esg, marker='^', label='UNH', ax=ax)
sns.lineplot(x=years, y='G-Score', data=mrk_annual_esg, marker='^', label='MRK', ax=ax)
ax.legend()
plt.show()
```

**Governance Performance (G-Score):**
   - LLY had the highest G-Score in 2020 (12.59), followed closely by MRK (11.07) and JNJ (11.44), suggesting better governance performance compared to UNH (5.96).
   - However, LLY's and JNJ's G-Scores declined over the years, with LLY dropping from 12.59 in 2020 to 9.14 in 2023, and JNJ dropping from 11.44 in 2020 to 8.03 in 2023.
   - MRK's G-Score also declined but remained relatively stable, ranging from 8.9 to 11.07.
   - UNH had the lowest G-Scores among the companies, ranging from 5.59 to 5.96, suggesting weaker governance performance.


```python
# Total Score Comparison
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Total Score Comparison")
ax.set_xlabel("Year")
ax.set_ylabel("Score")
sns.lineplot(x=years, y='Total-Score', data=lly_annual_esg, marker='^', label='LLY', ax=ax)
sns.lineplot(x=years, y='Total-Score', data=jnj_annual_esg, marker='^', label='JNJ', ax=ax)
sns.lineplot(x=years, y='Total-Score', data=unh_annual_esg, marker='^', label='UNH', ax=ax)
sns.lineplot(x=years, y='Total-Score', data=mrk_annual_esg, marker='^', label='MRK', ax=ax)
ax.legend()
plt.show()
```

**Overall ESG Performance (Total-Score):**
   - LLY had the highest overall ESG score in 2020 (34.19), followed by JNJ (33.92), MRK (27.22), and UNH (20.88).
   - However, LLY's and JNJ's overall ESG scores have been declining over the years, with LLY dropping from 34.19 in 2020 to 24.28 in 2023, and JNJ dropping from 33.92 in 2020 to 23.98 in 2023.
   - UNH and MRK also experienced declines in their overall ESG scores, but their scores were lower compared to LLY and JNJ throughout the period.
