This code will resample the original jnj_esg DataFrame to an annual frequency using the 'A' rule (for Annual), and it will calculate the mean value for each year.

```python
import pandas as pd
annual_esg = jnj_esg.resample('A').mean()
annual_esg.index = annual_esg.index.year
annual_esg
```

```python
import pandas as pd

# Assuming you have the four DataFrames already loaded
dataframes = {
    'LLY': lly_esg,
    'JNJ': jnj_esg,
    'UNH': unh_esg,
    'MRK': mrk_esg
}

# Create an empty dictionary to store the annual ESG DataFrames
annual_esg_dfs = {}

# Loop through each DataFrame
for ticker, df in dataframes.items():
    # Convert the index to monthly periods
    df.index = df.index.to_period('M')
    
    # Resample to annual frequency and take the mean
    annual_df = df.resample('A').mean()
    
    # Change the index to include only the year
    annual_df.index = annual_df.index.year
    
    # Store the annual DataFrame in the dictionary
    annual_esg_dfs[ticker] = annual_df

# Print the annual ESG DataFrames
for ticker, annual_df in annual_esg_dfs.items():
    print(f"Annual ESG Data for {ticker}:")
    print(annual_df)
    print("\n")
```
